```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
dataset = pd.read_excel("Superstore_USA.xlsx")
```

```python
dataset.head(2)
```

```python
dataset.shape
```

```python
# CHECKING_NULL_VAL
```

```python
dataset.isnull().sum()
```

# CATEGORY

```python
category_counts = dataset['Category'].value_counts()

plt.figure(figsize=(6, 4))
sns.histplot(category_counts, kde=False, bins=10)
plt.title('Distribution of Counts per Category')
plt.xlabel('Number of Orders')
plt.ylabel('Number of Categories')
plt.show()
```

# CUSTOMER SEGMENT

```python
customer_order_counts = dataset['Customer ID'].value_counts()

plt.figure(figsize=(6, 4))
sns.histplot(customer_order_counts, kde=False, bins=30)
plt.title('Distribution of Order Counts per Customer')
plt.xlabel('Number of Orders')
plt.ylabel('Number of Customers')
plt.show()
```

# SHIPPING MODE

```python
dataset['Ship Mode'].value_counts()
```

```python
x = dataset['Ship Mode'].value_counts().index
y = dataset['Ship Mode'].value_counts().values
```

```python
plt.figure(figsize=(5, 4))
plt.pie(y,labels=x,startangle= 90,autopct="%0.2f%%")
plt.legend(loc=3)
plt.show()
```

```python
plt.figure(figsize=(5, 4))
sns.countplot(x="Ship Mode", data=dataset, hue="Region")
plt.show()
```

# ORDER ID

```python
plt.figure(figsize=(5, 4))
sns.countplot(x="Order ID", data=dataset)
plt.xticks(rotation=90)  # Rotate x-axis labels if there are many unique order IDs
plt.title("Count of Order ID")
plt.savefig("Count of Order ID")
plt.show()
```
